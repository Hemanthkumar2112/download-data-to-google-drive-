```
def is_downloadable(url):
    h = requests.head(url, allow_redirects=True)
    header = h.headers
    content_type = header.get('content-type')
    if 'text' in content_type.lower():
        return False
    if 'html' in content_type.lower():
        return False
    return True
```

```
import tarfile


fname = "/content/drive/MyDrive/datavalidation/fer2013.tar.gz"
paths = '/content/drive/My Drive/face'

if fname.endswith("tar.gz"):
    print("file is in .tar.gz")
    tar = tarfile.open(fname, "r:gz")
    tar.extractall(path= paths)
    tar.close()
    print("Done...")

elif fname.endswith("tar"):
    print("file is in .tar")
    tar = tarfile.open(fname, "r:")
    tar.extractall(path = paths)
    tar.close()
    print("Done")
```
